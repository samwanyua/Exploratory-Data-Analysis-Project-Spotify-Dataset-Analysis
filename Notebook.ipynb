## Data Analysis

### Top 5 popular artists

```python
top_5_artists = df.groupby("Artist").count().sort_values(by="Song_title", ascending=False)["Song_title"][:5]
top_5_artists
```

```python
top_5_artists.plot.barh()
plt.show()
```

### Top Loudest tracks

```python
top_5_loudest_tracks = df[["Loudness","Song_title"]].sort_values(by="Loudness", ascending=True)[:5]
top_5_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="Song_title", y="Loudness", data=top_5_loudest_tracks)
plt.title("Top 5 Loudest Tracks")
plt.xticks(rotation=45)
plt.show()
```

### Artist with the most danceability song

```python
top_5_artist_dancable_songs = df[["Danceability", "Song_title", "Artist"]].sort_values(by="Danceability", ascending=False)[:5]
top_5_artist_dancable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="Danceability", y="Artist", data=top_5_artist_dancable_songs)
plt.title("Artist with the most danceable song")
plt.show()
```

### Top 10 instrumentalness tracks

```python
top_10_instrumental_tracks = df[["Instrumentalness", "Song_title", "Artist"]].sort_values(by="Instrumentalness", ascending=False)[:5]
top_10_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.title("Top 10 Instrumentalness Tracks")
plt.pie(x="Instrumentalness", data=top_10_instrumental_tracks, autopct="%1.2f%%")
plt.show()
```

### Multiple feature plots

```python
interested_feature_cols = ["Tempo", "Loudness", "Acousticness", "Danceability", "Duration_ms", "Energy", "Instrumentalness", "Liveness", "Speechiness", "Valence"]
```

```python
# plot histogram
for feature_col in interested_feature_cols:
    positive_data = df[df["Target"] == 1][feature_col]
    negative_data = df[df["Target"] == 0][feature_col]

    plt.figure(figsize=(12,7))
    sns.displot(positive_data, bins=30, label= "Positive", color="green")
    sns.displot(negative_data, bins=30, label="Negative", color="red")

    plt.legend(loc="upper right")
    plt.title(f"Positive and Negative Histogram for {feature_col}")
    plt.show()

```
